```python
import math

def calcular_altura_maxima(velocidad_inicial, angulo_grados):
    """
    Calcula la altura máxima de un proyectil.

    Parámetros:
    velocidad_inicial (float): La rapidez inicial en m/s.
    angulo_grados (float): El ángulo de lanzamiento en grados.

    Retorna:
    La altura máxima en metros.
    """
    # Constante de gravedad (m/s^2)
    g = 9.81

    # Convertir el ángulo de grados a radianes
    angulo_radianes = math.radians(angulo_grados)

    # Calcular el seno del ángulo
    seno_angulo = math.sin(angulo_radianes)

    # Aplicar la fórmula: h_max = (v0^2 * sin^2(theta)) / (2 * g)
    altura_maxima = (velocidad_inicial**2 * seno_angulo**2) / (2 * g)

    return altura_maxima

# Programa principal
print("Calculadora de Altura Máxima de un Proyectil")

try:
    # Solicitar datos al usuario
    v0 = float(input("Ingresa la rapidez inicial (m/s): "))
    angulo = float(input("Ingresa el ángulo de lanzamiento (grados): "))

    # Realizar el cálculo
    h_max = calcular_altura_maxima(v0, angulo)

    # Mostrar el resultado
    print(f"\nLa altura máxima alcanzada por el proyectil es de: {h_max:.2f} metros.")

except ValueError:
    print("Por favor, ingresa únicamente valores numéricos.")
```
